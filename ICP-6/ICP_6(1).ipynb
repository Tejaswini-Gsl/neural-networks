```python
import pandas
from keras.models import Sequential
from keras.layers import Dense, Activation

from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python

dataset = pd.read_csv("diabetes.csv", header=None).values

X_train, X_test, Y_train, Y_test = train_test_split(dataset[:,0:8], dataset[:,8],
                                                    test_size=0.25, random_state=87)
np.random.seed(155)
my_first_nn = Sequential() # create model
my_first_nn.add(Dense(20, input_dim=8, activation='relu')) # hidden layer
my_first_nn.add(Dense(1, activation='sigmoid')) # output layer
my_first_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
my_first_nn_fitted = my_first_nn.fit(X_train, Y_train, epochs=100,
                                     initial_epoch=0)
print(my_first_nn.summary())
print(my_first_nn.evaluate(X_test, Y_test))
```

2. Change the data source to Breast
Cancer dataset * available in the source code folder and make required
changes. Report accuracy of the model.


```python
bc_df = pd.read_csv("breastcancer.csv", header=[0])

bc_df.head()
```

```python
bc_df.isnull().sum()
```

```python
bc_df = bc_df.drop(bc_df.columns[0], axis=1)  # Drop ID column
bc_df = bc_df.dropna(axis=1) # drop with nan
```

```python
bc_df.shape
```

```python
bc_df['diagnosis'] = bc_df.iloc[:, 0].map({'M': 1, 'B': 0})
bc_df.head()
```

```python
X = bc_df.iloc[:, 1:].values  # Features
y = bc_df.iloc[:, 0].values  # Labels
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,y,
                                                    test_size=0.25, random_state=87)
np.random.seed(155)
breast_cancer_nn = Sequential() # create model
breast_cancer_nn.add(Dense(20, input_dim=X_train.shape[1], activation='relu')) # hidden layer
breast_cancer_nn.add(Dense(10, activation='relu')) # hidden layer
breast_cancer_nn.add(Dense(1, activation='sigmoid')) # output layer
breast_cancer_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
breast_cancer_nn_fitted = breast_cancer_nn.fit(X_train, Y_train, epochs=100,
                                     initial_epoch=0)
breast_cancer_nn.summary()
print(breast_cancer_nn.evaluate(X_test, Y_test))
```
